* ID = Household ID
* RAUMTYP = Population of Region, 1 = >100,000, 5 = <5,000
* PLZ = Postcode
* GKZ = State, 8 = BW
* LAGE = Location of apartment, 1 = city center, 6 = rural
* HHTYP = Type of Household, 1/2 = 1-2 p., 3/4 = +3 p.
* HHGRO = Number of people in household
* HHGRONEW = Number of people in household age >10
* P0_10 = Of that, Number of children under 10
* PKWHH = Number of cars
* PARKSTR1 = Type of parking lot, 2 = private Garage, 1 = Street/Public, for Car 1
* PARKSTR2 = Type of parking lot, 2 = private Garage, 1 = Street/Public, for Car 2
* PARKSTR3 = Type of parking lot, 2 = private Garage, 1 = Street/Public, for Car 3
* GEWHHWO = Projection Factor

# Processing Data

```python
# Sort by ID in TANK data set
df_TANK.sort_values(by=['IDHH'], inplace=True)
```

```python
import numpy as np

df_TANK['BATTERY'] = np.nan
df_TANK['POWER'] = np.nan
df_TANK['ENERGY'] = np.nan
df_TANK

# Add segment specific attributes: battery, power, and energy consumption
# Segment 1-4: Small, 5-8: Medium, 9-13: Large
def f_BATTERY(row):
    if row['SEGMENT'] == 1.0 or row['SEGMENT'] == 2.0 or row['SEGMENT'] == 3.0 or row['SEGMENT'] == 4.0:
        val = 41
    elif row['SEGMENT'] == 5.0 or row['SEGMENT'] == 6.0 or row['SEGMENT'] == 7.0 or row['SEGMENT'] == 8.0:
        val = 50
    else:
        val = 64
    return val

def f_POWER(row):
    if row['SEGMENT'] == 1.0 or row['SEGMENT'] == 2.0 or row['SEGMENT'] == 3.0 or row['SEGMENT'] == 4.0:
        val = 80
    elif row['SEGMENT'] == 5.0 or row['SEGMENT'] == 6.0 or row['SEGMENT'] == 7.0 or row['SEGMENT'] == 8.0:
        val = 239
    else:
        val = 150
    return val

def f_ENERGY(row):
    if row['SEGMENT'] == 1.0 or row['SEGMENT'] == 2.0 or row['SEGMENT'] == 3.0 or row['SEGMENT'] == 4.0:
        val = 17.2
    elif row['SEGMENT'] == 5.0 or row['SEGMENT'] == 6.0 or row['SEGMENT'] == 7.0 or row['SEGMENT'] == 8.0:
        val = 14.7
    else:
        val = 16.2
    return val

df_TANK['BATTERY'] = df_TANK.apply(f_BATTERY, axis=1)
df_TANK['POWER'] = df_TANK.apply(f_POWER, axis=1)
df_TANK['ENERGY'] = df_TANK.apply(f_ENERGY, axis=1)

print(df_TANK.head(100))
```

```python
# ID display option
pd.options.display.float_format = '{:20,.2f}'.format
df_W['ID'] = df_W['ID'].astype(np.int64)

# Add PKW number and driver count to each route
df_W['PKWNR'] = np.nan

# For all households that have PKWHH = 1, PKWNR = 1
def f_PKWNR(row):
    idhh = row[0]
    value = df_TANK.loc[df_TANK['IDHH'] == idhh, 'PKWHH']
    if value.empty:
        val = -1
    else:
        value_pkwnr = value.values[0]
        if value_pkwnr == 1:
            val = 1
        else:
            val = 0
    return val
    
df_W['PKWNR'] = df_W.apply(f_PKWNR, axis=1)
df_W.head(100)
```

# PKWNR Editing

```python
# Summary
df_HH_copy = df_HH #.....------------ 1525 households
#df_HH_copy
df_TANK_copy = df_TANK #------------- 1666 logs
#df_TANK_copy
df_W_copy = df_W # ------------------- 33318 trips --- 1495 unique ids/households
#df_W_copy

ids = df_W_copy['ID']

# Get all unique ids from W data set and loop through each household to assign vehicles to persons
ids_unique = ids.drop_duplicates()

pkw_values = []

for i in ids_unique:
    test_id = i
    print(f"ID: {test_id}")
    test = df_W_copy.loc[df_W_copy['ID'] == test_id]
    
    # Get person count and vehicle count
    persnrs = df_W_copy.loc[df_W_copy['ID'] == test_id, 'PERSNR'].drop_duplicates()
    perscount = persnrs.count()
    pkwnrs = df_TANK_copy.loc[df_TANK_copy['IDHH'] == test_id, 'PKWNR']
    if(pkwnrs.empty):
        for j in test.values:
            pkw_values.append(0)
        continue
    pkwcount = df_TANK_copy.loc[df_TANK_copy['IDHH'] == test_id, 'PKWHH'].values[0]
    print(persnrs)
    print(perscount)
    print(pkwnrs)
    print(pkwcount)
    
    for j in test.values:
        # If perscount == 1, then all persons have the same car
        if perscount == 1:
            pkw_values.append(pkwnrs.values[0])
            continue

        # If perscount == 2 and 2 or more vehicles: assign vehicles in order
        if (perscount == 2 and pkwcount == 2) or (perscount == 2 and pkwcount == 3) or (perscount == 2 and pkwcount == 4) or (perscount == 2 and pkwcount == 5) or (perscount == 2 and pkwcount == 6):
            try:
                if j[1] == persnrs.values[0]:
                    pkw_values.append(pkwnrs.values[0])
                else:
                    pkw_values.append(pkwnrs.values[1])
            except:
                # Error: 2 persons but only 1 car, take first car for second person
                pkw_values.append(pkwnrs.values[0])
                
                continue
            continue
            
        #if perscount == 3 and 3 or more vehicles: assign vehicles in order
        if (perscount == 3 and pkwcount == 3) or (perscount == 3 and pkwcount == 4) or (perscount == 3 and pkwcount == 5) or (perscount == 3 and pkwcount == 6):
            try:
                if j[1] == persnrs.values[0]:
                    pkw_values.append(pkwnrs.values[0])
                elif j[1] == persnrs.values[1]:
                    pkw_values.append(pkwnrs.values[1])
                else:
                    pkw_values.append(pkwnrs.values[2])
            except:
                # Error: 3 persons but < 3 cars, take first car for second and third person
                pkw_values.append(pkwnrs.values[0])
                continue
            continue
            
        #if perscount == 4 and 4 or more vehicles: assign vehicles in order
        if (perscount == 4 and pkwcount == 4) or (perscount == 4 and pkwcount == 5) or (perscount == 4 and pkwcount == 6):
            try:
                if j[1] == persnrs.values[0]:
                    pkw_values.append(pkwnrs.values[0])
                elif j[1] == persnrs.values[1]:
                    pkw_values.append(pkwnrs.values[1])
                elif j[1] == persnrs.values[2]:
                    pkw_values.append(pkwnrs.values[2])
                else:
                    pkw_values.append(pkwnrs.values[3])
            except:
                # Error: 4 persons but < 4 cars, take first car for every other person
                pkw_values.append(-4)
                continue
            continue
            
        #if perscount == 5 and 5 or more vehicles: assign vehicles in order
        if (perscount == 5 and pkwcount == 5) or (perscount == 5 and pkwcount == 6):
            try:
                if j[1] == persnrs.values[0]:
                    pkw_values.append(pkwnrs.values[0])
                elif j[1] == persnrs.values[1]:
                    pkw_values.append(pkwnrs.values[1])
                elif j[1] == persnrs.values[2]:
                    pkw_values.append(pkwnrs.values[2])
                elif j[1] == persnrs.values[3]:
                    pkw_values.append(pkwnrs.values[3])
                else:
                    pkw_values.append(pkwnrs.values[4])
            except:
                # Error: 5 persons but < 5 cars, take first car for every other person
                pkw_values.append(-5)
                continue
            continue
            
        #if perscount == 6 and 6 or more vehicles: assign vehicles in order
        if (perscount == 6 and pkwcount == 6):
            try:
                if j[1] == persnrs.values[0]:
                    pkw_values.append(pkwnrs.values[0])
                elif j[1] == persnrs.values[1]:
                    pkw_values.append(pkwnrs.values[1])
                elif j[1] == persnrs.values[2]:
                    pkw_values.append(pkwnrs.values[2])
                elif j[1] == persnrs.values[3]:
                    pkw_values.append(pkwnrs.values[3])
                elif j[1] == persnrs.values[4]:
                    pkw_values.append(pkwnrs.values[4])
                else:
                    pkw_values.append(pkwnrs.values[5])
            except:
                # Error: 6 persons but < 6 cars, take first car for every other person
                pkw_values.append(-6)
                continue
            continue
            
        pkw_values.append(1)

        
print(pkw_values)
```

```python
df_W_energy = df_W
# Add energy consumption to W data set
df_W_energy['ENERGYCONS'] = np.nan

# Add energy consumption according to consumption rate from assigned vehicle 
def f_ENERGYCONS(row):
    idhh = row[0]
    pkwnr = row[13]
    if pkwnr == 0:
        consumption = 0.16
        val = consumption * row[10]
    else:  
        consumption = df_TANK.loc[(df_TANK['IDHH'] == idhh) & (df_TANK['PKWNR'] == pkwnr), 'ENERGY']
        cons = consumption.values[0]/100
        val = cons * row[10]
    return val

    
df_W_energy['ENERGYCONS'] = df_W_energy.apply(f_ENERGYCONS, axis=1)
df_W_energy.head(100)
```
