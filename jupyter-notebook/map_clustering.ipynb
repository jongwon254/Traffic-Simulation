```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns; sns.set()
import csv
```

## Work Cluster

```python
# Import data file with geocoordinates of all work charging stations
df = pd.read_csv('all_work.csv')
df.head(20)
```

```python
# Run the kmeans clustering algorithm 
kmeans = KMeans(n_clusters = 350, init ='k-means++')
# Compute k-means clustering
kmeans.fit(df[df.columns[1:3]]) 
df['cluster_label'] = kmeans.fit_predict(df[df.columns[1:3]])
# Coordinates of cluster centers
centers = kmeans.cluster_centers_ 
# Labels of each point
labels = kmeans.predict(df[df.columns[1:3]]) 
df.head(10)
```

```python
# Coordinates of centroids
print(centers)
```

```python
# Visualization of clusters
df.plot.scatter(x = 'lat', y = 'lon', c=labels, s=10, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=20, alpha=0.5)
```

```python
# Export clustered data
df.to_excel('cluster_work.xlsx', index=None, header = True)
df.to_csv('cluster_work.csv', index=None, header = True)
```
